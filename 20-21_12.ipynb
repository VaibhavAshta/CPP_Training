```python
#We need sorted array for Binary Search

def binary_search(array, low, high, x):
    while low<= high:
        mid = low + (high - low)//2

        if array[mid] == x:
            del array[mid]
        elif array[mid] < x:
            low = mid + 1
        else:
            high = mid - 1
array = [3,4,5,6,7,8,9]
x     = 5
result = binary_search(array, 0, len(array) - 1, x)
print("Element found at index",result,"number is ",x)
array
```

```python
def bubble_sort(a):
    for i in range(len(a)):
        for j in range(0,len(a) -i-1):
            if a[j] > a[j+1]:
                (a[j], a[j+1]) = (a[j+1], a[j])
d = [-2,45,0,11,-9]
bubble_sort(d)
print("The sorted array in Ascending order is ")
print(d)
```
