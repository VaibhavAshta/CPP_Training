```python
import time
def add(num):
    for n in num:#[2,4,6,8,9]
        time.sleep(0.9)
        print("\tAddition= ", n+n)
        print()
def mul(num):
    for n in num:
        time.sleep(0.95)
        print("\tMultipilcation= ",n*n)
        print()
def sub(num):
    for n in num:
        time.sleep(1)
        print("\tSubraction= ",n-n)
        print()
num=[2,4,6,8,9]        
starttime = time.time()
t1=Thread(target=add,args=(num,))
t2=Thread(target=mul,args=(num,))
t3=Thread(target=sub,args=(num,))
t1.start()
t2.start()
t3.start()
t1.join()
t2.join()
t3.join()
print("Total time =",time.time()-starttime)
```
